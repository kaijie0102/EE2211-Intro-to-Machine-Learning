part(b): 
1. compute for each outcome (i,j,k): the value of $ijk/2$ and $(i + j + k)$.
2. compare and only include those (i,j,k) which fit our conditional

```python
d={(i,j,k):((i*j*k)/2 > i+j+k) for i in range(1,7) 
                               for j in range(1,7)
                                    for k in range(1,7)} 

# (1,1,1) : False
d
```

```python
dinv = defaultdict(list)
for vector,boolean_value in d.items(): 
    dinv[boolean_value].append(vector)
```

```python
# all outcomes are equiprobable with 1/6^3 probability 
X={i:len(j)/6.0**3 for i,j in dinv.items() }
print(X)
```

```python
from scipy import stats
# define constants 
mu = 30 # mean = 30
sigma = 1.8 # standard deviation = 1.8 
lower_bound = 28 # lower bound = 28 
upper_bound = 33 # upper bound = 33
```

recall that the CDF(x) gives the cumulative probability from - inf to x


```python
## calculate probabilities
# probability from x= infty to lower bound 
p_lower = stats.norm.cdf(lower_bound,mu,sigma) 
# probability from x= infty to upper bound 
p_upper = stats.norm.cdf(upper_bound,mu,sigma) 
# probability of the interval
prob = (p_upper) - (p_lower)
```

```python
# print the results 
print(f'Normal distribution: mean = {mu}, std dev = {sigma} \n') 
print(f'Probability of resistance between {lower_bound} and {upper_bound}:')
print(f'{round(prob*100,1)}%') 
```
