### final code for solution #1

```python
#final code
import csv
from matplotlib import pyplot as plt 
file_name = "GovernmentExpenditureonEducation.csv"
# from os import listdir
# from os.path import isfile, join, expanduser
yearList = []
expenditureList = []

# "data wrangling" portion
with open(file_name, mode = 'r') as csvfile: # file is accessible within this segment
    data = csv.reader(csvfile)
    
    for row in data:
        if row[0] != 'year':
            yearList.append(int(row[0]))
            expenditureList.append(int(row[1]))

# this is where we use matplotlib to present the data
plt.plot(yearList, expenditureList)
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.title('Education Expenditure')
plt.show()
```
